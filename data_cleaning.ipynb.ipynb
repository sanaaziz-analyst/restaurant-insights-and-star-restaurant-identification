```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_") 
#column names validation
```

```python
df.head()
```

```python
df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
#strip spaces from all text column
```

```python
numeric_cols = ["restaurant_id", "country_code", "longitude", "latitude",
                "average_cost_for_two", "price_range", "aggregate_rating", "votes"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    # converted numeric columns
```

```python
df.isnull().sum() # missing values check
```

```python
df=df.drop_duplicates() # removing duplicates
```

```python
df=df.fillna("Unknown") #replacing missing values
```

```python
df.to_csv("cleaned_marketing_project.csv", index=False)
```
